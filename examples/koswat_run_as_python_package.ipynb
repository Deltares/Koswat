# Use case: Koswat runs in a script as a Python package

In case you did not install koswat yet, you will have to do so via command line, for this purpose you can check our [Koswat installation documentation](https://deltares.github.io/Koswat/installation.html).

## Simple script

This simplified script will run the model kept in our `basic_case` directory and generate a log in a new `koswat_log_output` directory.

```python
# Import the required packages.
from pathlib import Path
from koswat.koswat_handler import KoswatHandler
```

```python
# Locate our model to run
_examples_dir = Path(__file__).parent
_basic_case_dir = _examples_dir.joinpath("basic_case")
assert _basic_case_dir.is_dir()

_config_file = _basic_case_dir.joinpath("koswat_general.json")
assert _config_file.is_file()
```

```python
# Create a results directory
_log_output_dir = _examples_dir.joinpath("koswat_log_output")
_log_output_dir.mkdir()
```

```python
# Run model
with KoswatHandler(_log_output_dir) as _handler:
    _handler.run_analysis(_config_file)
```
